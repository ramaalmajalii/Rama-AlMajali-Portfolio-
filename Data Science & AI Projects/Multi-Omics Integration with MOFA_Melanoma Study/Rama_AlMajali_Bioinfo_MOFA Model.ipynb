```python
import pandas as pd
import numpy as np
from mofapy2.run.entry_point import entry_point
ent = entry_point()
```

```python
clinical_scaled_df = pd.read_csv("clinical_processed (3).csv")      # Has SAMPLE_ID
clinical_scaled_df = clinical_scaled_df.drop(columns=["Unnamed: 0"])

expression_scaled_df = pd.read_csv("data_mrna_seq_fpkm.csv")  # Has SAMPLE_ID
# 2. Set gene IDs as index
expression_scaled_df = expression_scaled_df.set_index("Entrez_Gene_Id")

# 3. Transpose so that samples are rows
expression_scaled_df = expression_scaled_df.T.reset_index()
expression_scaled_df = expression_scaled_df.rename(columns={"index": "SAMPLE_ID"})
```

```python
clinical_long = clinical_scaled_df.copy()
clinical_long["group"] = "-"
clinical_long = clinical_long.melt(id_vars=["SAMPLE_ID", "group"],
                                   var_name="feature",
                                   value_name="value")
clinical_long = clinical_long.rename(columns={"SAMPLE_ID": "sample"})
clinical_long["view"] = "Clinical"
```

```python
expression_long = expression_scaled_df.copy()
expression_long["group"] = "-"
expression_long = expression_long.melt(id_vars=["SAMPLE_ID", "group"],
                                       var_name="feature",
                                       value_name="value")
expression_long = expression_long.rename(columns={"SAMPLE_ID": "sample"})
expression_long["view"] = "Expression"
```

```python
mofa_input_df = pd.concat([clinical_long, expression_long], ignore_index=True)
```

```python
# Set data options
ent.set_data_options(scale_views=False)
```

```python
# Check duplicates (optional: display a few)
duplicates = mofa_input_df[mofa_input_df.duplicated(subset=["group", "view", "feature", "sample"], keep=False)]
print(f"{len(duplicates)} duplicates found")

# Drop duplicates
mofa_input_df = mofa_input_df.drop_duplicates(subset=["group", "view", "feature", "sample"])
```

```python
mofa_input_df["group"] = mofa_input_df["group"].astype(str)
mofa_input_df["sample"] = mofa_input_df["sample"].astype(str)
mofa_input_df["feature"] = mofa_input_df["feature"].astype(str)
mofa_input_df["view"] = mofa_input_df["view"].astype(str)
```

```python
mofa_input_df["value"] = pd.to_numeric(mofa_input_df["value"], errors="coerce")
```

```python
#Load data into MOFA
ent.set_data_df(mofa_input_df, likelihoods=["gaussian", "gaussian"])
```

```python
# 4. Set model options
ent.set_model_options(
    factors=10,
    spikeslab_weights=True,
    ard_weights=True
)
```

```python
# 5. Set training options
ent.set_train_options(
    convergence_mode="fast",
    dropR2=0.001,
    gpu_mode=False,  # set True if you have GPU
    seed=42
)
```

```python
# 6. Build and run model
ent.build()
ent.run()
```

```python
# 7. Save model
ent.save("MOFA_model_final.hdf5")
```
