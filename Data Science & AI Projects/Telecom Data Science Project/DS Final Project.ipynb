```python
#I first noticed there is a strong relation between 'Tenure Category' & 'Customer Tenure'
#Here i decided to select the missing values in 'Tenure Category' with their corresponding value in 'Customer Tenure'
missing = df.loc[df['Tenure Category'].isnull(), 'Customer Tenure'].value_counts()#it counts how when each tenure value appears where the 'Tenure Category' is missing.
missing 
#From this we can gather that all the missing values in tenure category correlate to 88/87/25 in customer tenure
```

```python
#Now to know what to do with the prior information i wanted to check the range of values in 'Customer Tenure' using the max and min and their corresponding values in 'Tenure Category'
min_tenure = df.groupby('Tenure Category')['Customer Tenure'].min()
max_tenure = df.groupby('Tenure Category')['Customer Tenure'].max()

# Create a new DataFrame directly using these values
tenure_ranges = pd.DataFrame({
    'Tenure Category': min_tenure.index,
    'Min Tenure': min_tenure.values,
    'Max Tenure': max_tenure.values,
})
tenure_ranges
```

```python
#After getting the ranges previously we notice that each category is related to a range of numbers, and that is how i am gonna fill it
#Filling 'Tensure Category'
def categorize(val):
    if 0 <= val <= 88:
        return 'Short'
    elif 89 <= val <= 119:
        return 'Medium'
    elif 120 <= val <= 153:
        return 'Long'
df.loc[df['Tenure Category'].isna(), 'Tenure Category'] = df['Customer Tenure'].apply(lambda val: categorize(val))
df.isnull().sum()
```

# Modeling & Visualization


```python
#Splitting the data.
# Dropping columns that won't be used as features
columns_to_drop = ['CO_ID', 'POS_MSISDN', 'MSISDN','STATUS','deactive year','deactive month','deactive day','Customer Tenure',
                   'Tenure Category_Long','Tenure Category_Medium','Tenure Category_Short']
X = df.drop(columns_to_drop, axis=1)  # Features
y = df['STATUS']# Target variable
```

```python
#Making a dataframe to display iterations.
results = pd.DataFrame(columns=['Model', 'Random_State', 'Accuracy','Precision','Recall','F1_score'])
results
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split

import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y)
X_train = np.ascontiguousarray(X_train)
X_test = np.ascontiguousarray(X_test)
iterations = 30

k=3
knn = KNeighborsClassifier(n_neighbors=k)

for i in range(iterations):
    # Training the KNeighborsClassifier
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = i)
    X_train = np.ascontiguousarray(X_train)
    X_test = np.ascontiguousarray(X_test)
    
    # Training the KNeighborsClassifier
    knn.fit(X_train, y_train)

    # Making predictions on the test set
    y_pred = knn.predict(X_test)

    # Calculate the evaluation metrics of the model
    knn_accuracy = accuracy_score(y_test, y_pred)
    knn_precision = precision_score(y_test, y_pred, average='weighted')
    knn_recall = recall_score(y_test, y_pred, average='weighted')
    knn_f1_score = f1_score(y_test, y_pred, average='weighted')

    results.loc[len(results)] = ['kNN', i+1, knn_accuracy, knn_precision, knn_recall, knn_f1_score]
results
```

```python
knn_acc = []
knn_pre = []
knn_rec = []
knn_f1 = []
knn_acc.append(knn_accuracy)
knn_rec.append(knn_recall)
knn_pre.append(knn_precision)
knn_f1.append(knn_f1_score)
```

```python
data = [knn_acc,knn_rec,knn_pre, knn_f1]
plt.boxplot(data,labels = ['Accuracy_KNN','Recall_KNN','Precision_KNN','F1-KNN'])
plt.show()
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

# Create Decision Tree classifer object
dt_classifier = DecisionTreeClassifier()

for i in range(iterations):
    # Split the data into train and test sets with a different random state each time
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)

    # Train the Decision Tree Classifier
    dt_classifier.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = dt_classifier.predict(X_test)

    # Print metrics for each iteration
    DT_accuracy = accuracy_score(y_test, y_pred)
    DT_precision = precision_score(y_test, y_pred, average='weighted')
    DT_recall = recall_score(y_test, y_pred, average='weighted')
    DT_f1_score = f1_score(y_test, y_pred, average='weighted')

    results.loc[len(results)] = ['Decision Tree', i+1, DT_accuracy, DT_precision, DT_recall, DT_f1_score]
results
```

```python
dt_acc = []
dt_pre = []
dt_rec = []
dt_f1 = []
dt_acc.append(DT_accuracy)
dt_rec.append(DT_recall)
dt_pre.append(DT_precision)
dt_f1.append(DT_f1_score)
```

```python
data = [dt_acc,dt_pre,dt_rec,dt_f1]

plt.boxplot(data,labels = ['Accuracy_DT','Precision_DT','TRecall_DT','F1-DT'])
plt.show()
```

# Gaussian NB

```python
from sklearn.naive_bayes import GaussianNB

for i in range(iterations):
    # Split the data into train and test sets with a different random state each time
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=i)

    # Create Gaussian Naive Bayes classifier object
    NB = GaussianNB()

    # Train Gaussian Naive Bayes Classifier
    NB.fit(X_train, y_train)

    # Make predictions on the test set
    y_pred = NB.predict(X_test)

    # Calculate metrics for each iteration
    NB_accuracy = accuracy_score(y_test, y_pred)
    NB_precision = precision_score(y_test, y_pred, average='weighted')
    NB_recall = recall_score(y_test, y_pred, average='weighted')
    NB_f1_score = f1_score(y_test, y_pred, average='weighted')

    
    results.loc[len(results)] = ['Gaussian Naive Bayes', i+1, NB_accuracy, NB_precision, NB_recall, NB_f1_score]
results
```

```python
nb_acc = []
nb_pre = []
nb_rec = []
nb_f1 = []
nb_acc.append(NB_accuracy)
nb_rec.append(NB_recall)
nb_pre.append(NB_precision)
nb_f1.append(NB_f1_score)

    
```
