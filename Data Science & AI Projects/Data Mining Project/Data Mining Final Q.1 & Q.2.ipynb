# *Text Mining*

### Importing the necessary libraries 

```python
import nltk
import numpy as np
import pandas as pd 
from nltk.stem import PorterStemmer
import string
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.spatial.distance import euclidean, cityblock, cosine
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
#Reading the CSV file
df = pd.read_csv(r"C:\Users\ramam\Desktop\Final_students_data.csv")
df
```

## Question 1 

```python
df_bios = df.copy()
```

## Tokenization

```python
df_bios ['Bio Tokens'] = df_bios ['Bio'].apply(nltk.word_tokenize)
print(df_bios [['Name', 'Bio Tokens']])
```

## Removing stop words

```python
stop_words = set(stopwords.words('english'))
df_bios ['Bio Tokens Cleaned'] = df_bios ['Bio Tokens'].apply(lambda tokens: [word for word in tokens if word.lower() not in stop_words])
print(df_bios [['Name', 'Bio Tokens Cleaned']])
```

## Removing Punctutation 

```python
df_bios ['Bio Tokens Cleaned No Punctuation'] = df_bios ['Bio Tokens Cleaned'].apply(lambda tokens: [word for word in tokens if word not in string.punctuation])
print(df_bios [['Name', 'Bio Tokens Cleaned No Punctuation']])
```

## Lowercasing 

```python
df_bios ['Bio Tokens Lowercase'] = df_bios ['Bio Tokens Cleaned No Punctuation'].apply(lambda tokens: [word.lower() for word in tokens])
print(df_bios [['Name', 'Bio Tokens Lowercase']])
```

## Stemming

```python
stemmer = PorterStemmer()
df_bios ['Bio Tokens Stemmed'] = df_bios ['Bio Tokens Lowercase'].apply(lambda tokens: [stemmer.stem(word) for word in tokens])
print(df_bios[ ['Bio Tokens Stemmed']])
```

## Repeating the same steps for the 5 queries 

```python
query1 = "Tell me about someone interested in machine learning"
tokens1 = nltk.word_tokenize(query1)
tokens1 = [word for word in tokens1 if word.lower() not in stop_words]
tokens1 = [word for word in tokens1 if word not in string.punctuation]
tokens1 = [word.lower() for word in tokens1]
tokens1 = [stemmer.stem(word) for word in tokens1]

query2 = "Who is skilled in software development and programming languages?"
tokens2 = nltk.word_tokenize(query2)
tokens2 = [word for word in tokens2 if word.lower() not in stop_words]
tokens2 = [word for word in tokens2 if word not in string.punctuation]
tokens2 = [word.lower() for word in tokens2]
tokens2 = [stemmer.stem(word) for word in tokens2]

query3 = "Looking for a data scientist with experience in R and Python"
tokens3 = nltk.word_tokenize(query3)
tokens3 = [word for word in tokens3 if word.lower() not in stop_words]
tokens3 = [word for word in tokens3 if word not in string.punctuation]
tokens3 = [word.lower() for word in tokens3]
tokens3 = [stemmer.stem(word) for word in tokens3]

query4 = "Find someone with a cybersecurity background"
tokens4 = nltk.word_tokenize(query4)
tokens4 = [word for word in tokens4 if word.lower() not in stop_words]
tokens4 = [word for word in tokens4 if word not in string.punctuation]
tokens4 = [word.lower() for word in tokens4]
tokens4 = [stemmer.stem(word) for word in tokens4]

query5 = "Who is experienced in business intelligence and data visualization?"
tokens5 = nltk.word_tokenize(query5)
tokens5 = [word for word in tokens5 if word.lower() not in stop_words]
tokens5 = [word for word in tokens5 if word not in string.punctuation]
tokens5 = [word.lower() for word in tokens5]
tokens5 = [stemmer.stem(word) for word in tokens5]

print("Processed Query 1:", tokens1)
print("Processed Query 2:", tokens2)
print("Processed Query 3:", tokens3)
print("Processed Query 4:", tokens4)
print("Processed Query 5:", tokens5)
```

## Calculating TF-IDF

```python
clean_queries = [' '.join(tokens1), ' '.join(tokens2), ' '.join(tokens3), ' '.join(tokens4),' '.join(tokens5)]
```

```python
bios_joined = [' '.join(bio) for bio in df_bios['Bio Tokens Stemmed']]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

for q_idx, query in enumerate(clean_queries, 1):  
    documents = bios_joined 
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(documents)

    tfidf_query = vectorizer.transform([query]) 

    cosine_similarities = cosine_similarity(tfidf_query, tfidf_matrix)

    results = []
    for idx, sim in enumerate(cosine_similarities[0]):
        results.append({
            'Bio': f'Doc{idx + 1}',
            'Bio Content': documents[idx],
            'Cosine Similarity': sim
        })

    results = pd.DataFrame(results)

    results['Rank'] = results['Cosine Similarity'].rank(ascending=False, method='dense')

    results = results.sort_values(by='Rank')

    print(f"Results for Query {q_idx}:")
    print(results[['Rank', 'Bio', 'Bio Content', 'Cosine Similarity']]) 
    print() 
```

## Question 2 

```python
#We have to clean the user entered query
def preprocess_query(query):
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(query)
    tokens = [word for word in tokens if word.lower() not in stop_words]  
    tokens = [word for word in tokens if word not in string.punctuation]  
    tokens = [word.lower() for word in tokens]  
    stemmer = PorterStemmer()
    tokens_stemmed = [stemmer.stem(word) for word in tokens]  
    return ' '.join(tokens_stemmed)  

def rank_bios_with_std(query, bios_joined):
    cleaned_query = preprocess_query(query)
    documents = bios_joined + [cleaned_query]  
    
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(documents)

    tfidf_docs = tfidf_matrix[:-1]  
    tfidf_query = tfidf_matrix[-1] 

    cosine_similarities = cosine_similarity(tfidf_query, tfidf_docs)

    ranked_bios = sorted(zip(cosine_similarities[0], bios_joined), reverse=True, key=lambda x: x[0])

    cosine_scores = [score for score, bio in ranked_bios]
    mean_score = np.mean(cosine_scores)
    std_score = np.std(cosine_scores)

    print(f"Mean Cosine Similarity: {mean_score}")
    print(f"Standard Deviation of Cosine Similarities: {std_score}")

    threshold = mean_score + std_score 
    print(f"Threshold (Mean + Std): {threshold}")

    filtered_bios = [bio for score, bio in ranked_bios if score >= threshold]
    print(f"\nResults for the query: '{query}'")
    relevant_bios_found = False
    for rank, (score, bio) in enumerate(ranked_bios, 1):
        if score >= threshold and score > 0: 
            relevant_bios_found = True
            print(f"Rank {rank}: {bio} (Score: {score})")

    if not relevant_bios_found:
        print("No relevant bios.")

user_query = input("Enter your query: ")
rank_bios_with_std(user_query, bios_joined)
```
